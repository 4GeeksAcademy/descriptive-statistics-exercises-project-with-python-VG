### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import numpy as np
import matplotlib.pyplot as plt

mu=5
sigma=0.5

s = np.random.normal(mu, sigma, 100)


count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
````

````python
chi =np.random.chisquare(3,100)

plt.figure(figsize=(8, 5))
plt.hist(chi, bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title('Distribucin Chi-cuadrado con 3 grados de libertad')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
````

````python
import statistics as stats
from scipy.stats import skew
from scipy.stats import kurtosis
import pandas as pd

df = pd.DataFrame(s,columns=['Normal'])
print(df.describe())
df_chi=pd.DataFrame(chi, columns=["Chi-square"])
print(df_chi.describe())

````
